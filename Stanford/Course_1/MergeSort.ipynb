```python
class MergeSort:

    @staticmethod
    def merge(left, right):
        print('merging')
        result = []
        while left and right:
            if left[0] <= right[0]:
                result.append(left[0])
                left = left[1:]
            else:
                result.append(right[0])
                right = right[1:]
        # If there are remaining elements in left, add them to result
        while left:
            result.append(left[0])
            left = left[1:]
        # If there are remaining elements in right, add them to result
        while right:
            result.append(right[0])
            right = right[1:]
        return result

    @staticmethod
    def sort(x):
        # Base case: a list with 0 or 1 elements is already sorted
        if len(x) <= 1:
            print('basecase')
            return x
        # Find the midpoint of the list
        midpoint = len(x) // 2
        # Divide the list into two halves
        left = x[:midpoint]
        right = x[midpoint:]
        # Recursively sort both halves
        left = MergeSort.sort(left)
        print(left)
        right = MergeSort.sort(right)
        print('left and right: ',left,right)
        # Merge the two sorted halves
        return MergeSort.merge(left, right)

# Example usage
arr = [38, 27, 43, 3, 9, 82, 10]
sorted_arr = MergeSort.sort(arr)
print(sorted_arr)  # Output: [3, 9, 10, 27, 38, 43, 82]
```

```python
def merge_and_count(arr, temp_arr, left, mid, right):
    i = left    # Starting index for left subarray
    j = mid + 1 # Starting index for right subarray
    k = left    # Starting index to be sorted
    inv_count = 0

    # Conditions are checked to ensure that i doesn't exceed mid and j doesn't exceed right
    while i <= mid and j <= right:
        if arr[i] <= arr[j]:
            temp_arr[k] = arr[i]
            i += 1
        else:
            # There are mid - i inversions, because all the remaining elements in the left subarray
            # (arr[i], arr[i+1], ..., arr[mid]) are greater than arr[j]
            temp_arr[k] = arr[j]
            inv_count += (mid-i + 1)
            j += 1
        k += 1

    # Copy the remaining elements of left subarray, if any
    while i <= mid:
        temp_arr[k] = arr[i]
        i += 1
        k += 1

    # Copy the remaining elements of right subarray, if any
    while j <= right:
        temp_arr[k] = arr[j]
        j += 1
        k += 1

    # Copy the sorted subarray into Original array
    for i in range(left, right + 1):
        arr[i] = temp_arr[i]

    return inv_count

def merge_sort_and_count(arr, temp_arr, left, right):
    inv_count = 0
    if left < right:
        mid = (left + right)//2

        inv_count += merge_sort_and_count(arr, temp_arr, left, mid)
        inv_count += merge_sort_and_count(arr, temp_arr, mid + 1, right)

        inv_count += merge_and_count(arr, temp_arr, left, mid, right)

    return inv_count

def count_inversions(arr):
    n = len(arr)
    temp_arr = [0]*n
    return merge_sort_and_count(arr, temp_arr, 0, n - 1)

# Read the file and convert it to a list of integers
with open('IntegerArray.txt', 'r') as file:
    array = [int(line.strip()) for line in file]

# Get the number of inversions
inversion_count = count_inversions(array)
print(inversion_count)
```
