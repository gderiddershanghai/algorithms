```python
# Finds the minimal divisor>1 of the given integer m>1
def min_divisor(m):
    for d in range(2, m + 1):
        if m % d == 0:
            return d
        # optimization:
        if d * d > m:
            return m


def is_prime(m):
    return m == min_divisor(m)


def primes_list(n):
    lst = []
    boundary = 2
    # primes < boundary are in lst
    while len(lst) < n:
        if is_prime(boundary):
            lst.append(boundary)
        boundary += 1

    return lst


print('The first ten primes:')
primes = primes_list(250)
```

```python
import matplotlib.pyplot as plt

def min_divisor(m):
    for d in range(2, m + 1):
        if m % d == 0:
            return d
        if d * d > m:
            return m

def is_prime(m):
    return m == min_divisor(m)

def primes_list(n):
    lst = []
    boundary = 2
    while len(lst) < n:
        if is_prime(boundary):
            lst.append(boundary)
        boundary += 1
    return lst

# Generate a list of the first 100 primes
n = 10000
primes = primes_list(n)

# Calculate differences between consecutive primes
differences = []
for i in range(len(primes) - 1):
    diff = primes[i + 1] - primes[i]
    differences.append(diff)

# Plot the differences
plt.plot(differences, marker='o')
plt.title('Differences Between Consecutive Primes')
plt.xlabel('Index')
plt.ylabel('Difference')
plt.grid(True)
plt.show()
```

```python
plt.hist(differences, bins=range(1, max(differences) + 1), edgecolor='black')
plt.title('Histogram of Differences Between Consecutive Primes')
plt.xlabel('Difference')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
