```python
from collections import deque

def bfs(graph, start):
    visited = set()
    queue = deque([start])
    
    while queue:
        vertex = queue.popleft()
        if vertex not in visited:
            visited.add(vertex)
            print(vertex)  # This can be replaced with any operation you want to perform on the node
            
            # Add all unvisited neighbors to the queue
            for neighbor in graph[vertex]:
                if neighbor not in visited:
                    queue.append(neighbor)
                    
    return visited
```

```python
def dfs_recursive(graph, start, visited=None):
    if visited is None:
        visited = set()
    visited.add(start)
    print(f'start: {start}',visited)  # This can be replaced with any operation you want to perform on the node
    
    for neighbor in graph[start]:
        if neighbor not in visited:
            dfs_recursive(graph, neighbor, visited)

    return visited
```

```python
def dfs_iterative(graph, start):
    visited = set()
    stack = [start]
    
    while stack:
        vertex = stack.pop()
        if vertex not in visited:
            visited.add(vertex)
            print(vertex)  # This can be replaced with any operation you want to perform on the node
            
            # Add all unvisited neighbors to the stack
            for neighbor in graph[vertex]:
                if neighbor not in visited:
                    stack.append(neighbor)
                    
    return visited
```

```python
from collections import deque

def bfs(graph, start, visited):
    queue = deque([start])
    component = set()
    
    while queue:
        node = queue.popleft()
        if node not in visited:
            visited.add(node)
            component.add(node)
            for neighbor in graph[node]:
                if neighbor not in visited:
                    queue.append(neighbor)
    
    return component

def find_connected_components_bfs(graph):
    visited = set()
    components = []
    
    for node in graph:
        if node not in visited:
            component = bfs(graph, node, visited)
            components.append(component)
    
    return components
```

```python
# Example graph
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E'],
    'G': ['H'],
    'H': ['G']
}

components = find_connected_components_bfs(graph)
for i, component in enumerate(components):
    print(f"Component {i+1}: {component}")
```

```python
import networkx as nx
import matplotlib.pyplot as plt


def steiner_tree(graph, terminals):
    # Create a subgraph induced by the terminals
    subgraph = graph.subgraph(terminals)
    
    # Find the minimum spanning tree of the subgraph
    mst = nx.minimum_spanning_tree(subgraph)
    
    # Add Steiner points to the graph
    steiner_graph = graph.copy()
    for u, v, data in mst.edges(data=True):
        steiner_graph.add_edge(u, v, weight=data['weight'])
    
    return steiner_graph
```

```python
# Example graph
graph = nx.Graph()
edges = [
    (0, 1, 4), (0, 7, 8), (1, 2, 8), (1, 7, 11),
    (2, 3, 7), (2, 8, 2), (2, 5, 4), (3, 4, 9),
    (3, 5, 14), (4, 5, 10), (5, 6, 2), (6, 7, 1),
    (6, 8, 6), (7, 8, 7)
]
graph.add_weighted_edges_from(edges)

# Terminals to be connected
terminals = [0, 2, 4, 6]

# Find the approximate Steiner tree
steiner_tree_graph = steiner_tree(graph, terminals)

# Print the edges in the Steiner tree
print("Edges in the Steiner Tree:")
for u, v, data in steiner_tree_graph.edges(data=True):
    print(f"({u}, {v}, {data['weight']})")
```

```python
# Plot the original graph
pos = nx.spring_layout(graph)  # positions for all nodes
plt.figure(figsize=(10, 8))

# Draw the original graph in light grey
# nx.draw(graph, pos, with_labels=True, node_color='lightgrey', edge_color='lightgrey', node_size=700, font_size=16, font_weight='bold')
# nx.draw_networkx_edge_labels(graph, pos, edge_labels={(u, v): d['weight'] for u, v, d in graph.edges(data=True)}, font_color='grey')

# Draw the Steiner tree in blue
nx.draw(steiner_tree_graph, pos, with_labels=True, node_color='red', edge_color='blue', node_size=700, font_size=16, font_weight='bold')
nx.draw_networkx_edge_labels(steiner_tree_graph, pos, edge_labels={(u, v): d['weight'] for u, v, d in steiner_tree_graph.edges(data=True)}, font_color='blue')

plt.title('Approximate Steiner Tree')
plt.show()
```

```python
import numpy as np
import networkx as nx

# Distance matrix D
D = np.array([[0, 3, 6, 5],
              [3, 0, 5, 6],
              [6, 5, 0, 3],
              [5, 6, 3, 0]])

# Create a graph from the distance matrix
G = nx.from_numpy_array(D)
print(G)
# Find the minimum spanning tree (MST)
mst = nx.algorithms.tree.mst.minimum_spanning_tree(G)

# Print the edges in the MST
for edge in mst.edges(data=True):
    print(edge)

# Calculate the total weight of the MST
total_weight = sum(weight['weight'] for _, _, weight in mst.edges(data=True))
print(f"Total weight of MST: {total_weight}")
```
